```python
import csv

def extract_csv_to_nested_list(file_path):
    nested_list = []
    with open(file_path, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        for row in csv_reader:
            nested_list.append(row)
    return nested_list

# Example usage:
file_path = '/content/drive/MyDrive/Colab Notebooks/Projects/Datasets/covid_abstracts.csv'  # Replace 'data.csv' with the path to your CSV file
nested_list = extract_csv_to_nested_list(file_path)
print(nested_list)
```
